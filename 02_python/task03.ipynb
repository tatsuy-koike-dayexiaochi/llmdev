```python
def calculate_interest(amount: float, rate: float) -> float:
    return amount * rate

def task03(principal: int, annual_interest_rate_percent: float):
    if principal < 0:
        return "元金は0円以上で入力してください。"
    if annual_interest_rate_percent < 0:
        return "年利は0%以上で入力してください。"

    rate = annual_interest_rate_percent / 100
    final_amount = float(principal)

    # 繰り返し構文を使用して、10年間の金額を計算します。
    for i in range(10):
        final_amount += calculate_interest(final_amount, rate)
    
    return f"10年後の金額: {round(final_amount, 2)}"

# ユーザーが元金（初期投資額）を入力します。入力には input() 関数を使用してください。
principal = int(input("元金を入力してください: "))
annual_interest_rate_percent = float(input("年利を入力してください: "))

result = task03(principal, annual_interest_rate_percent)
print(result)
```

```python
# 元金が1000円、年利が5%の場合
result = task03(1000, 5)
expected = "10年後の金額: 1628.89"

assert result == expected
print(result)
```
