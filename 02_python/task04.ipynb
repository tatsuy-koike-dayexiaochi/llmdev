```python
def task04(input_a: str, input_b: str):
    try:
        numerator = float(input_a)
        denominator = float(input_b)
    except ValueError:
        # 数値の代わりに文字を入力した場合は「エラー: 数値を入力してください」と表示します。
        return "エラー: 数値を入力してください"

    # 割る数が0の場合は「エラー: 0で割ることはできません」と表示します。
    if denominator == 0:
        return "エラー: 0で割ることはできません"

    # 計算が正常に完了した場合、結果を表示してください。
    return f"計算結果: {round(numerator/denominator, 2)}"


# ユーザーに2つの数値を入力させ
input_a = input("割られる数を入力してください: ")
input_b = input("割る数を入力してください: ")


result = task04(input_a, input_b)
print(result)
```

```python
# 1つ目の数を 100、2つ目の数を 10 と入力した場合
result = task04(100, 10)
expected = "計算結果: 10.0"

assert result == expected
print(result)
```

```python
# 1つ目の数を 100、2つ目の数を 0 と入力した場合
result = task04(100, 0)
expected = "エラー: 0で割ることはできません"

assert result == expected
print(result)
```

```python
# 1つ目の数を a と入力した場合
result = task04("a", 0)
expected = "エラー: 数値を入力してください"

assert result == expected
print(result)
```

```python
# 1つ目の数を 100、2つ目の数を b と入力した場合
result = task04(100, "b")
expected = "エラー: 数値を入力してください"

assert result == expected
print(result)
```
