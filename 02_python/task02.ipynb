```python
def get_tax_rate() -> float:
    return 0.1

def calculate_price_incl_tax(price_excl_tax: int) -> int:
    return int(price_excl_tax * (1 + get_tax_rate())) # 仕様不明の為、一旦切り捨て

def calculate_shipping_fee(price_incl_tax: int) -> int:
    shipping_fee = 350
    if (price_incl_tax >= 2000) :
        shipping_fee = 0
    return shipping_fee

def task02(price_excl_tax: int):
    if price_excl_tax < 0:
        return "税抜価格は0円以上で入力してください。"
    # 税込価格が2000円以上の場合は送料無料
    # 2000円未満の場合は350円
    # 消費税率は10%
    price_incl_tax = calculate_price_incl_tax(price_excl_tax)
    shipping_fee = calculate_shipping_fee(price_incl_tax)
    total_price = price_incl_tax + shipping_fee
    shipping_fee_str = "無料" if shipping_fee == 0 else f"{shipping_fee}円"
    
    return f"""送料は{shipping_fee_str}です
送料込税込価格は{total_price}円です"""

result = task02(int(input("税抜価格を入力してください: ")))
print(result)
```

```python
# 税抜価格が2000円の場合
result = task02(2000)
expected = """送料は無料です
送料込税込価格は2200円です"""

assert result == expected
print(result)
```

```python
# 税抜価格が1500円の場合
result = task02(1500)
expected = """送料は350円です
送料込税込価格は2000円です"""

assert result == expected
print(result)
```
