```python
# クラスを作成
class BankAccount:
    # 残高を管理するための属性を定義
    def __init__(self, balance):
        # 初期残高を指定してインスタンス化できるようにします。
        if balance < 0:
            raise ValueError("エラー: 正しい金額を入力してください")
        
        self._balance = balance
    
    @property
    def balance(self):
        return self._balance

    # 残高属性は外部から直接アクセスできないようにしてください。
    # -> セッターの用意なし
    
    # 指定された金額を残高に追加します
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
        else:
            # 預け入れる金額が0より小さい場合はエラー
            print("エラー: 正しい金額を入力してください")
    
    # 指定された金額を残高から引き出します
    def withdraw(self, amount):
        if amount > 0 and self._balance >= amount :
            # 引き出し可能な場合のみ残高を減らし
            self._balance -= amount
        else:
            # 残高不足や0以下の金額を指定した場合はエラー
            print("エラー: 残高不足または無効な金額です。")
    
    def get_balance(self):
        return self.balance
```

```python
# 呼び出し側の仕様

# 初期残高1000円で BankAccount クラスのインスタンスを作成
bank = BankAccount(1000)

# 500円を預け入れ
bank.deposit(500)

# 200円を引き出し
bank.withdraw(200)

# 現在の残高を表示（1300 と表示される）
print(bank.get_balance())

# 1500円を引き出そうとする（エラー: 残高不足または無効な金額です。）
bank.withdraw(1500)
```

```python
# 預け入れ時に無効な金額（-500など）を指定した場合
bank = BankAccount(1000)
bank.deposit(-500)
```

```python
# 0以下の金額を指定した場合
bank = BankAccount(1000)
bank.withdraw(0)
```

```python
# 0以下の金額を指定した場合
bank = BankAccount(1000)
bank.withdraw(-500)
```

```python
# balanceにセットできないことの確認
bank = BankAccount(1000)
bank.balance = 5000
```
