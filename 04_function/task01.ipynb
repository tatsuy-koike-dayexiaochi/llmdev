```python
# バリデーションの結果をブール値で返す関数を作成してください。

# !ユーザー名は5文字以上15文字以内
def is_valid_len(text):
    is_valid = 5 <= len(text) <= 15
    if not is_valid:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
    return is_valid

# ユーザー名に空白は使えない
def is_valid_no_space(text):
    is_valid = " " not in text
    if not is_valid:
        print("エラー: ユーザー名に空白を含めることはできません")
    return is_valid

# ユーザー名は英数字（アルファベットと数字）のみ使用可能
def is_valid_alphanumeric(text):
    is_valid = text.isalnum()
    if not is_valid:
        print("エラー: ユーザー名は英数字のみ使用できます")
    return is_valid

def task01(input_name):
    if (
        is_valid_len(input_name)
        and is_valid_no_space(input_name)
        and is_valid_alphanumeric(input_name)
    ):
        print("ユーザー名は有効です")
    else:
        print("不正なユーザー名です")

# 文字列を受け取り
input_name = input("ユーザー名を入力してください:")
task01(input_name)
```

```python
# JohnDoe123 と入力した場合
# バリデーション違反時にprintしている為、戻り値比較が難しいのでprintを監視するテストにしました
import io
from contextlib import redirect_stdout

buffer = io.StringIO()

with redirect_stdout(buffer):
    task01('JohnDoe123')

result = buffer.getvalue()
expected = "ユーザー名は有効です\n"

assert result == expected
print(result)
```

```python
# JD と入力した場合
import io
from contextlib import redirect_stdout

buffer = io.StringIO()

with redirect_stdout(buffer):
    task01('JD')

result = buffer.getvalue()
expected = """エラー: ユーザー名は5文字以上15文字以内で入力してください
不正なユーザー名です
"""

assert result == expected
print(result)
```
