```python
def task01(grades: list):
    # 成績の平均点を計算します。
    avg_score = sum(grades) / len(grades)
    # 成績の最高点と最低点を求めます。
    max_score = max(grades)
    min_score = min(grades)
    # 成績が平均以上の学生数をカウントします。
    avg_score_cnt = len([score for score in grades if score >= avg_score])
    
    return f"""平均点: {round(avg_score, 1)}
最高点: {max_score}
最低点: {min_score}
平均以上の学生数: {avg_score_cnt}"""

grades = [85, 90, 78, 92, 88]
result = task01(grades)
print(result)
```

```python
# 成績リストが [85, 90, 78, 92, 88] の場合
grades = [85, 90, 78, 92, 88]
result = task01(grades)
expected = """平均点: 86.6
最高点: 92
最低点: 78
平均以上の学生数: 3"""

assert result == expected
print(result)
```
