```python
def print_alphabet_cnt(alphabet_cnt_items):
    return "\n".join(
        f"{alphabet}が{count}個ありました"
        for alphabet, count in alphabet_cnt_items
    )

def task02(input_list: list):
    # 入力された英単語をABC順に並べ替えて表示します。
    input_list.sort()
    # すべての英単語に含まれるアルファベットごとの数も表示します。
    alphabet_cnt = {}
    for word in input_list:
        for alphabet in word:
            alphabet_cnt[alphabet] = alphabet_cnt.get(alphabet, 0) + 1
    
    return f"""入力した英単語：{input_list}
{print_alphabet_cnt(sorted(alphabet_cnt.items()))}"""


input_list = []
while True:
    # 英単語の入力を受け付けます。
    input_str = input("英単語を入力してください：")

    # 未入力でリターン（エンター）キーを押されるまで、英単語の受付を繰り返します。
    if input_str :
        input_list.append(input_str)
        continue

    break
result = task02(input_list)
print(result)
```
